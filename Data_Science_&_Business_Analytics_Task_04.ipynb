```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'day','gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude','longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill','nwound':'Wound'},inplace=True)
df.columns.values
```

```python
data = df[['Year','Month','day','Country','State','Region','City','latitude','longitude','Attacktype','kill','Group','Wound','Targettype','Weapon','motive']]
data
```

```python
data.isnull().sum()
```

```python
data['Wound']=data['Wound'].fillna(0)
data['kill']=data['kill'].fillna(0)
```

```python
data.info()
```

Descriptive Statistics

```python
data.describe()
```

```python
year=data['Year'].unique()
years_count=df['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(15,10))
sns.barplot(x=year,y=years_count,palette="tab10")
plt.xticks(rotation=50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks per year',fontsize=20)
plt.title('Attacks in years from 1970 to 2017',fontsize=30)
plt.show()
```

We can see a drastic increase in the number of attacks from the year 2013, 2014 is the peak year of terror attacks

```python
crosstab(data.Year,data.Region).plot(kind='area',stacked=False,figsize=(15,10))
plt.title('Region Wise terrorist Attacks for Each Year',fontsize=20)
plt.ylabel('Number of attacks',fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.show()
```

```python
attack=data.Country.value_counts()[:10]
attack
```

```python
data.Group.value_counts()[1:10]
```

```python
dx=data[['Year','kill']].groupby(['Year']).sum()
fig,ax2=plt.subplots(figsize=(18,10))
dx.plot(kind='bar',alpha=0.7,ax=ax2)
plt.xticks(rotation=50)
plt.title("People Died Due to Attack",fontsize=25)
plt.ylabel("Number of killed people",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side=ax2.spines["top"]
top_side.set_visible(False)
```

The Maximum Number of deaths have occured in the year 2014

```python
plt.figure(figsize=(40,10))
plt.xticks(rotation=90)
plt.title('Attacks')
sns.countplot(x=data.Country)
plt.show()
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation=50)
plt.xlabel('City',fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("Top 10 Most Affected City",fontsize=20)
plt.show()
```

```python
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation=60)
plt.title('Number of killed',fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
data[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='red')
plt.xticks(rotation=50)
plt.title("Number of Wounded",fontsize=20)
plt.ylabel("Number of People",fontsize=15)
plt.xlabel("Attack type",fontsize=15)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')

plt.xlabel('Terrorist group Name',fontsize=15)
plt.ylabel("Attack Number",fontsize=15)
plt.title("Top 10 Terrorist group attack",fontsize=20)
plt.show()
```

```python
data[['Group','kill']].groupby(["Group"],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='red')

plt.title('Top 10 terrorist group attack',fontsize=20)
plt.ylabel('terrorist group name',fontsize=15)
plt.xlabel('No of Killed people',fontsize=15)
plt.show()
```
